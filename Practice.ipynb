```python
# Lists
My_List = ['a', 'b', 'c', 'd']

# Adding a single element at the end
My_List.append('e')  # append() adds only one element
print(My_List)

# Adding multiple elements
lst = [1, 2, 3]

# append() adds the entire list as ONE element (nested list)
lst.append([4, 5, 6])
print(lst)
# Output: [1, 2, 3, [4, 5, 6]]

# extend() adds each element individually
lst = [1, 2, 3]
lst.extend([4, 5, 6])
print(lst)
# Output: [1, 2, 3, 4, 5, 6]

# Indexing and Slicing
print(My_List[0:])    # Entire list
print(My_List[1:3])   # Elements from index 1 to 2
print(lst[2:4])       # Elements from index 2 to 3

# Modifying an element
My_List[0] = 'new'
print(My_List)
# Output: ['new', 'b', 'c', 'd', 'e']

# fetching an index from a nested list 
l = [1, 2, 3, 4, [23,45,66],12, 64]
print (l[4][1])

# multiple nested lists 3
lst_1 = [1, 2, [3, 34, [5, 'krishna', 43, 11, ['apple', 'mango', 'banana']]]]
print(lst_1[2][2][1])
```

```python
# Dictionaries 

my_dict = {'k1' : [1, 3, 5]}

#indexing a dictionary 
list_2 = my_dict ['k1']
print (list_2)

list_2[0]

#Dictionary nested into another dictionary 
student = {"name": "John", "age": 20, "marks": { "math": 90, "science": 85, "english": 88 }}
list_3 = student ['marks']
print (list_3)
```

```python
# Tuples
t = (1, 2, 3, 4)
print(t)

l = [1, 2, 3, 4] # list (mutable)
l[0] = 'new'
print (l) #['new', 2, 3, 4] is the output

t[0] = 'next' # tuples being immutable throws a TypeError  
print (t)
```

```python
#Sets 
s ={1,1,1,2,2,2,3,3,3,4,4,5}
print (s)

# Sets can also be written as 
nums = set([1,2,3,4,5,3,2,5,3,5,6,8,1])
print (nums)

# Adding a value to set 
nums.add (17) 
print (nums)

# Removing a Value 
nums.remove (2)
print(nums)
```

```python
# Comparison operators
a = 2
b = 4

print (a == b)
print (a != b)
print (a > b)
print (a < b)
print (a >= b)
print (a <= b)

# Using keywords with comparison operators
print ((a > b) and (a == b))
print ((a < b) or (a == b))

```

```python
# If Elif and Else Statements 
if 1 >= 2 :
    print ('true')

else :
    print ('false')

# using elif as well 
if 1 == 2 :
    print ('first')
elif 3 == 3:
    print ('second')
else :
    print ('third')

```
