```python
# Lists
My_List = ['a', 'b', 'c', 'd']

# Adding a single element at the end
My_List.append('e')  # append() adds only one element
print(My_List)

# Adding multiple elements
lst = [1, 2, 3]

# append() adds the entire list as ONE element (nested list)
lst.append([4, 5, 6])
print(lst)
# Output: [1, 2, 3, [4, 5, 6]]

# extend() adds each element individually
lst = [1, 2, 3]
lst.extend([4, 5, 6])
print(lst)
# Output: [1, 2, 3, 4, 5, 6]

# Indexing and Slicing
print(My_List[0:])    # Entire list
print(My_List[1:3])   # Elements from index 1 to 2
print(lst[2:4])       # Elements from index 2 to 3

# Modifying an element
My_List[0] = 'new'
print(My_List)
# Output: ['new', 'b', 'c', 'd', 'e']

# fetching an index from a nested list 
l = [1, 2, 3, 4, [23,45,66],12, 64]
print (l[4][1])

# multiple nested lists 3
lst_1 = [1, 2, [3, 34, [5, 'krishna', 43, 11, ['apple', 'mango', 'banana']]]]
print(lst_1[2][2][1])
```
